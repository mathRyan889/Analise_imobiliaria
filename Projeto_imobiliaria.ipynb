# Qual o valor médio de aluguel por tipo de imovel

```python
dados['Valor'].mean() #Vai calcular a media de todos os valores da coluna valor (ainda não é o que queremos)
```

Precisamos calcular a media por tipo de imovel, para agrupar os tipos de imoveis vamos usar a função **Groupby** que agrupa tipos iguais em uma coluna

```python
dados.groupby('Tipo').mean(numeric_only=True)
```

```python
dados.groupby('Tipo')['Valor'].mean() #como especificamos que queremos a coluna valor (uma coluna númerica), eliminamos a necessidade da função numeric_only=True
```

```python
dados.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
```

Vamos colocar esse data frame dentro de uma variavel para facilitar a criação de graficos

```python
df_preco_tipo = dados.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
df_preco_tipo.plot(kind='barh', figsize=(14,10), color='purple')
```

é possivel ver uma diferença discrepante no valor medio de alguns imoveis em comparativa a outros, isso se deve por ter tanto imoveis comerciais quanto imoveis residenciais, para trabalhos futuros como a criação de modelos de ML isso não é interesse e pode gerar erros nos treinamentos de ml, por isso iremos separar esses dados do dataframe


```python
dados.Tipo.unique()
```

```python
imoveis_comerciais = ['Conjunto Comercial/Sala',
                      'Prédio Inteiro','Loja/Salão',
                      'Galpão/Depósito/Armazém',
                      'Casa Comercial', 'Terreno Padrão',
                      'Loja Shopping/ Ct Comercial','Box/Garagem',
                      'Chácara','Loteamento/Condomínio',
                      'Sítio','Pousada/Chalé','Hotel', 'Indústria']
```

agora devemos remover todos os imoveis presentes nessa lista do nosso dataframe, iremos utilizar a função **Query** que nos permite selecionar um conjuto especifico de dados sob determinadas condições

```python
dados.query('@imoveis_comerciais in Tipo') # Selecione as linhas onde estão os imoveis comerciais na coluna tipo
```

```python
dados.query('@imoveis_comerciais not in Tipo') #Selecione todas as linhas menos as dos imoveis comerciais
```

```python
df = dados.query('@imoveis_comerciais not in Tipo') #Selecione todas as linhas menos as dos imoveis comerciais
df.head()
```

```python
df.Tipo.unique()
```

Agora nosso grafico está exibindo dados bem menos discrepantes

```python
df_preco_tipo = df.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
df_preco_tipo.plot(kind='barh', figsize=(14,10), color='purple',xlabel='Valores',ylabel='Tipos de imovel')
```

# Qual o percentual de cada tipo de imovel na nossa base de dados



para acharmos o percentual dos imoveis iremos utilizar inicialmente a função value_counts() ele vai contar quantas vezes cada tipo de imovel aparece

```python
df.Tipo.value_counts()
```

para ele exibir esses números em percentual vamos usar no parametro de value_counts() o normalize = True

```python
df.Tipo.value_counts(normalize=True)
```

Agora vamos melhorar a visualização dos dados **to_frame() transforma os dados em um dataframe**

```python
df.Tipo.value_counts(normalize=True).to_frame().sort_values('Tipo') #criando um dataframe
```

```python
df_percentual_tipo = df.Tipo.value_counts(normalize=True).to_frame().sort_values('Tipo')
df_percentual_tipo.plot(kind='bar', figsize=(14,10), color='green',
                        xlabel='Tipos',ylabel='Percentual')
```

como podemos ver no grafico, o imovel do tipo apartamento representam cerca de 80% dos imoveis da nossa base de dados, portanto iremos trabalhar somente com eles daqui em diante, não faz sentindo manter os demais

```python
df.query('Tipo == "Apartamento"') #Da coluna Tipo selecione Apartamento
```

```python
df = df.query('Tipo == "Apartamento"')
df.head()
```

# Tratando e filtrando dados

# lidando com dados nulos
para tratar os dados nulos vamos usar no processo a função **isnull()**, ela irá criar um dataframe com valores booleanos, sendo false para valores que não são nulos e True para valores nulos

```python
df.isnull()
```

```python
df.isnull().sum() #Vai realizar a soma de quantos valores nulos existem por colunas
```
